```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from numpy.linalg import svd

# Create a 15x40 grid filled with 0s
grid = np.zeros((15, 40))

# Define the pattern based on the image
# Letter 'H'
grid[2:10, 1:3] = 1
grid[5:7, 3:5] = 1
grid[2:10, 5:7] = 1

# Letter 'E'
grid[3:11, 9:11] = 1
grid[3:5, 11:14] = 1
grid[6:8, 11:14] = 1
grid[9:11, 11:14] = 1

# Letter 'L'
grid[4:12, 16:18] = 1
grid[10:12, 18:21] = 1

# Letter 'L'
grid[5:13, 23:25] = 1
grid[11:13, 23:28] = 1

# Letter 'O'
grid[6:15, 30:32] = 1
grid[6:8, 32:34] = 1
grid[13:15, 32:34] = 1
grid[6:15, 34:36] = 1


# Create the plot
fig, ax = plt.subplots(figsize=(12, 4))

# Plot the grid
for i in range(15):
    for j in range(40):
        if grid[i, j] == 1:
            ax.plot(j, i, 'ko', markersize=8)

# Set axis properties to match the original
ax.set_xlim(-1, 41)
ax.set_ylim(-1, 16)
ax.invert_yaxis()
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('15x40 Grid Plot')
ax.grid(False)

plt.tight_layout()
plt.show()

# Print the grid as 0s and 1s
print("Grid representation (0s and 1s):")
print(grid.astype(int))

rank = np.linalg.matrix_rank(grid)
print("Rank of the matrix is:", rank)
```

```python
# Compute SVD
U, S, VT = svd(grid)

# Print singular values
print("Singular values:")
print(S)

# Plot singular values on linear scale
plt.figure(figsize=(8, 5))
plt.plot(S, 'o-', label='Singular values')
plt.xlabel('Index')
plt.ylabel('Singular value')
plt.title('Singular Values of Grid Matrix (Linear Scale)')
plt.grid(True, axis='y')
plt.legend()
plt.tight_layout()
plt.show()

# Plot singular values with a log scale
plt.figure(figsize=(8, 5))
plt.plot(S, "o-", label="Singular values")
plt.yscale("log")
plt.xlabel("Index")
plt.ylabel("Singular value (log scale)")
plt.title("Singular Values of Grid Matrix (Log Scale)")
plt.grid(True, which="both", axis="y")
plt.legend()
plt.tight_layout()
plt.show()
```

The rank of a matrix A is the maximum number of linearly independent rows or columns in A.  Considering a matrix's SVD, it is exactly the number of its nonzero singular values in 𝝨.  Numerically (with floating point arithmetic), singular values may never be exactly zero, but can at least be extremely small, so one might consider counting the singular values significantly larger than zero in order to estimate the rank.  

```python
#some tolerance for rank
tol = 1e-10
rank = np.sum(S > tol)
print(f"Numerical rank: {rank}")
```

```python
# Custom grayscale colormap (white to black, 5 shades)
custom_gray = ListedColormap([
    [1, 1, 1],
    [0.75, 0.75, 0.75],
    [0.5, 0.5, 0.5],
    [0.25, 0.25, 0.25],
    [0, 0, 0]
])

for i in range(1, rank+1):
    # Best rank-i approximation
    Bi = np.dot(U[:, :i], np.dot(np.diag(S[:i]), VT[:i, :]))

    # Unflip
    Bi_display = np.flipud(Bi[:-1, :-1])



# Prepare 3x4 subplots for grayscale colormap
nrows, ncols = 3, 4
fig_gray, axes_gray = plt.subplots(nrows, ncols, figsize=(18, 12))
fig_gray.suptitle('Best Rank-i Approximations (Unflipped, Custom Grayscale)', fontsize=16)

# Prepare 3x4 subplots for pseudocolor colormap
fig_pseudo, axes_pseudo = plt.subplots(nrows, ncols, figsize=(18, 12))
fig_pseudo.suptitle('Best Rank-i Approximations (Pseudocolor)', fontsize=16)

for idx in range(nrows * ncols):
    i = idx + 1
    if i > rank:
        axes_gray.flat[idx].axis('off')
        axes_pseudo.flat[idx].axis('off')
        continue
    Bi = np.dot(U[:, :i], np.dot(np.diag(S[:i]), VT[:i, :]))

    # Grayscale, MATLAB-style unflipped
    Bi_display = np.flipud(Bi[:-1, :-1])
    pcm_gray = axes_gray.flat[idx].pcolor(Bi_display, cmap=custom_gray, edgecolors='k', linewidth=0.1)
    axes_gray.flat[idx].set_title(f'Rank-{i}')
    axes_gray.flat[idx].axis('equal')
    axes_gray.flat[idx].axis('off')

    # Pseudocolor plot (no flip, invert y for MATLAB pcolor style)
    pcm_pseudo = axes_pseudo.flat[idx].pcolor(Bi[:-1, :-1], cmap='viridis')
    axes_pseudo.flat[idx].set_title(f'Rank-{i}')
    axes_pseudo.flat[idx].axis('equal')
    axes_pseudo.flat[idx].axis('off')
    axes_pseudo.flat[idx].invert_yaxis()

fig_gray.colorbar(pcm_gray, ax=axes_gray, orientation='vertical', fraction=0.03, pad=0.04, label='Value')
plt.tight_layout(rect=[0, 0, 1, 0.96])
fig_gray.show()

fig_pseudo.colorbar(pcm_pseudo, ax=axes_pseudo, orientation='vertical', fraction=0.03, pad=0.04, label='Value')
plt.tight_layout(rect=[0, 0, 1, 0.96])
fig_pseudo.show()
```
